```python
def parse_game_info(game):
    '''Given a game element, returns a dictionary of the relevant information.'''
    
    output = {}
    
    # find all details for this game
    game_details = game.findChildren()

    # 'POS' or current ranking of game
    output['pos'] = game_details[0].text
    # Game Title
    output['title'] = game_details[7].text.strip()
    # Console the game is on
    output['console'] = game_details[8].findChildren('img')[0]['alt']
    # Publisher
    output['publisher'] = game_details[10].text.strip()
    # Developer
    output['developer'] = game_details[11].text.strip()
    # VG Chartz Score
    output['vgc_score'] = game_details[12].text.strip()
    # Critic Score
    output['critic_score'] = game_details[13].text.strip()
    # User Score
    output['user_score'] = game_details[14].text.strip()
    # Total Shipped
    output['total_shipped'] = game_details[15].text.strip()
    # Total Sales
    output['total_sales'] = game_details[16].text.strip()
    # NA Sales
    output['na_sales'] = game_details[17].text.strip()
    # PAL Sales
    output['pal_sales'] = game_details[18].text.strip()
    # Japan Sales
    output['japan_sales'] = game_details[19].text.strip()
    # Other Sales
    output['other_sales'] = game_details[20].text.strip()
    # Release Date
    output['release_date'] = game_details[21].text.strip() 
    # Last Update
    output['last_update'] = game_details[22].text.strip()
    
    return output
```

```python
def parse_VCG_page(url):
    '''Given a url, returns a dataframe where each row is a video game from the url.'''
    # send page request and make soup
    response = get(url)
    soup = BeautifulSoup(response.text)
    
    # search soup for the 'tr' tag, which each game is under
    table_rows = soup.select("tr", {"style": "background-image:url(../imgs/chartBar_large.gif); height:70px"})
    
    # Only keep the rows with games
    games = []
    for row in range(1,len(table_rows)):
        if re.search("https://www.vgchartz.com/game/.+\d./", str(table_rows[row])) == None :
            continue
        else:
            games.append(table_rows[row])
    
    # Create a dataframe 
    df = pd.DataFrame([parse_game_info(game) for game in games])
    
    return df
```

```python
def get_games():
    '''
    Returns a dataframe of games
    '''
    # Make url a variable
    url = 'https://www.vgchartz.com/games/games.php?page=1&results=200&order=Sales&ownership=Both&direction=DESC&showtotalsales=1&shownasales=1&showpalsales=1&showjapansales=1&showothersales=1&showpublisher=1&showdeveloper=1&showreleasedate=1&showlastupdate=1&showvgchartzscore=1&showcriticscore=1&showuserscore=1&showshipped=1'
    
    # Create the dataframe
    df = pd.DataFrame()
    

    df = pd.concat([df, pd.DataFrame(parse_VCG_page(url))])
   
    # Reset index
    df = df.reset_index(drop=True)
    
    return df
```
